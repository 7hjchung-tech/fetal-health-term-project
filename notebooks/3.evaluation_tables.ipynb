```python
import pandas as pd
import matplotlib.pyplot as plt

binary_path = "../reports/tables/performance_ctg_binary.csv"
df_binary = pd.read_csv(binary_path)

multi_path = "../reports/tables/performance_ctg_multiclass.csv"
df_multi = pd.read_csv(multi_path)

feat_path = "../reports/tables/logistic_feature_importance_long.csv"
df_feat = pd.read_csv(feat_path)
```

이진 분류에서 모델별 주요 지표

```python
metrics = ["sensitivity", "specificity", "ppv", "npv", "auc"]

plot_df_bin = df_binary.set_index("model")[metrics].T

plt.figure(figsize=(8, 5))
plot_df_bin.plot(kind="bar")
plt.title("High-risk vs Normal: Model-wise Metrics")
plt.ylabel("Score")
plt.ylim(0, 1.05)
plt.xticks(rotation=0)
plt.legend(title="Model", bbox_to_anchor=(1.05, 1), loc="upper left")
plt.tight_layout()
plt.show()
```

3-class에서 클래스별 F1 score 비교(bar plot)

```python
f1_cols = ["f1_cls1", "f1_cls2", "f1_cls3"]
plot_df_multi = df_multi.set_index("model")[f1_cols].T

class_labels = ["1 (Normal)", "2 (Suspicious)", "3 (Pathologic)"]
plot_df_multi.index = class_labels

plt.figure(figsize=(8, 5))
plot_df_multi.plot(kind="bar")
plt.title("Multiclass Fetal Health: Per-class F1 Score by Model")
plt.ylabel("F1 Score")
plt.ylim(0, 1.05)
plt.xticks(rotation=0)
plt.legend(title="Model", bbox_to_anchor=(1.05, 1), loc="upper left")
plt.tight_layout()
plt.show()
```

병적(class=3)에서 feature odds ratio 상위 중요도(hbar)

```python
df_pathologic = df_feat[df_feat["class"] == 3].copy()


N = 10
df_top = df_pathologic.sort_values("abs_coef", ascending=False).head(N)


df_top = df_top.iloc[::-1]

plt.figure(figsize=(8, 5))
plt.barh(df_top["feature"], df_top["odds_ratio"])
plt.xlabel("Odds Ratio (Class 3 vs Others)")
plt.title("Logistic Regression (Class 3, Pathologic): Top Features by Odds Ratio")
plt.tight_layout()
plt.show()
```

NIPT-cfDNA_자연어처리 문서 분석

```python
nipt_path = "../data/NIPT-cfDNA_자연어문서데이터.csv"
df_nipt = pd.read_csv(nipt_path)
```

```python
df_main = df_nipt[df_nipt["subgroup"] == "all"].copy()
df_main = df_main[df_main["condition"].isin(["T21", "T18", "T13"])]

def map_method_group(m):
    if m == "standard_screening":
        return "Standard"
    else:
        return "cfDNA"  

df_main["method_group"] = df_main["method"].apply(map_method_group)

summary_nipt = (
    df_main
    .groupby(["condition", "method_group"])[["sens", "spec", "ppv", "npv"]]
    .median()
    .reset_index()
)

summary_nipt
```

```python
pivot_sens = summary_nipt.pivot(
    index="condition",
    columns="method_group",
    values="sens"
)

plt.figure(figsize=(6, 4))
pivot_sens.plot(kind="bar")
plt.title("Sensitivity by Condition and Method (NIPT)")
plt.ylabel("Sensitivity (%)")
plt.ylim(0, 105)
plt.xticks(rotation=0)
plt.legend(title="Method", loc="lower right")
plt.tight_layout()
plt.show()
```

```python
t21_summary = summary_nipt[summary_nipt["condition"] == "T21"].set_index("method_group")

ppv_npv_t21 = t21_summary[["ppv", "npv"]]

plt.figure(figsize=(6, 4))
ppv_npv_t21.plot(kind="bar")
plt.title("T21: PPV and NPV by Method")
plt.ylabel("Value (%)")
plt.ylim(0, 105)
plt.xticks(rotation=0)
plt.legend(title="Metric", loc="lower right")
plt.tight_layout()
plt.show()
```
